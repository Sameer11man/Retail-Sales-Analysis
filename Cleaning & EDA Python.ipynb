# Step 1: Load and Preview the Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data=pd.read_csv("C:/Users/samee/project code and ds/Data/Superstore.csv" , encoding='latin-1')
df=pd.DataFrame(data)
df

print(df.head())
print(df.describe())
print(df.info())
```

## Step 2: Data Cleaning

```python
# Remove duplicates

df.drop_duplicates()
df
```

```python
# Handle missing values'
print(df.isnull())
print(df.isnull().sum())# Check for nulls
```

```python
# Convert Order Date to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'],format='mixed')
df['Ship Date'] =pd.to_datetime(df['Ship Date'], format='mixed')
# Create new column for Month-Year
df['Month'] = df['Order Date'].dt.to_period('M')
print(df)
```

Step 3: Exploratory Analysis

```python
#Sales by region
#typ1
d1=df.groupby("Region")["Sales"].sum()
print(d1)
plt.bar(d1.index, d1.values, color="skyblue")
plt.xlabel("Region")
plt.ylabel("Sales")
plt.title("Sales by region")
plt.show()


#type2
df.groupby('Region')['Sales'].sum().plot(kind='bar', title='Total Sales by Region', color='skyblue')
#df.groupby('Region').agg({"Sales":"sum"}).plot(kind='bar', title='Total Sales by Region', color='skyblue')
plt.ylabel('Sales')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
#2. Profit by Category

df.groupby('Category')['Profit'].sum().plot(kind='barh', title='Profit by Category', color='salmon')
plt.xlabel('Profit')
plt.tight_layout()
plt.show()
```

```python
# 3. Monthly Sales Trend

Monthly_Sales=df.groupby("Month")["Sales"].sum()
# plt.plot(Monthly_Sales.index.astype(str), Monthly_Sales.values,marker="D")
Monthly_Sales.plot(marker="D", title='Monthly sale trend')
#plt.xticks(ticks=range(0,len(Monthly_Sales.index), 3), labels=Monthly_Sales.index[::2], rotation=45)
plt.xlabel("Month")
plt.ylabel('Sales')
plt.title('Monthly sale trend')
plt.tight_layout()
plt.show()
```

```python
# 4. Top 10 Products by Sales

d1=df.groupby("Sub-Category")["Sales"].sum().sort_values(ascending=False).head(10)
d1.plot(kind='bar' ,title='Top 10 Products by Sales',color='green')
plt.ylabel("Total Sales")
plt.xlabel("Top Products")
plt.tight_layout()
plt.show()
```

```python
#🔸 4. Profit vs Discount Scatter Plot

#plt.scatter( y=df["Discount"] , x=df["Profit"]) # data=df,
plt.figure(figsize=(8, 5))
sns.scatterplot(x='Discount', y='Profit', data=df ,marker="D" , hue="Sub-Category")
plt.title('Profit vs Discount')
plt.legend(bbox_to_anchor=(0.2,0,1.1,1))
plt.xlabel('Discount')
plt.ylabel('Profit')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#🔸 5. Heatmap of Correlation
#Goal: Understand relationships between numeric variables.plt.figure(figsize=(8, 6))
numeric_df = df[["Sales", "Profit", "Discount", "Quantity"]] 
plt.figure(figsize=(6, 4))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```

```python
# 🔸 6. Sales & Profit by Segment and Region
# Goal: Understand business performance across segments.

pivot=df.pivot_table(index="Region" ,columns ="Segment",values=["Profit","Sales"], aggfunc="sum")
pivot.plot(kind="bar",figsize=(10, 6), title="Profit by segment and region") #figsize=(10, 6),
plt.ylabel("Amount")
plt.xticks(rotation=0)
plt.legend(loc='best',bbox_to_anchor=(1,0,0.5,1))
plt.tight_layout()
plt.show()
```

```python
# 🔸 4. Negative Profit Transactions (Loss-Making Orders)
# Goal: Identify risky areas.

loss_orders = df[df['Profit'] < 0]
print("Loss-making orders:", loss_orders.shape[0])
loss_by_subcat = loss_orders['Sub-Category'].value_counts().head()
print(loss_by_subcat)
```

```python
# 🔸 5. Top Cities by Sales

top_cities = df.groupby('City')['Sales'].sum().sort_values(ascending=False).head(10)
top_cities.plot(kind='barh', title='Top 10 Cities by Sales', color='teal')
plt.xlabel("Sales")
plt.tight_layout()
plt.show()
```
